```python
import wikipedia
wpage = wikipedia.page("Hep c")
a = wpage.url
import requests
from bs4 import BeautifulSoup
page = requests.get(a)
# Create a BeautifulSoup object
soup = BeautifulSoup(page.text, 'html.parser')
catlist_list = soup.find(class_='mw-normal-catlinks')
categories = catlist_list.find_all('a')
a1=0
l=[]
Master = []
for x in categories:
    if(a1==0):
        a1=1
    else:
        print(x.text)
        l.append(x.text) 

        
def urlmaker(l,count):
    urlx = ""
    i=""
    for i in l:    
        Part1 = "https://en.wikipedia.org/wiki/Special:CategoryTree?target=Category%3A"
        Part2 = i
        Part3 = "&mode=categories&namespaces=&title=Special%3ACategoryTree"
        urlx = Part1+Part2+Part3
        treebuilder(urlx,count,i)

def treebuilder(urlx,count,i):    
    cattreepage = requests.get(urlx)
    soup = BeautifulSoup(cattreepage.text, 'html.parser')
    Pcatlist_list = soup.find(class_='CategoryTreeParents')
    Parentcategories = Pcatlist_list.find_all('a')
    
    pl=[]
    
    for x in Parentcategories:
        pl.append(x.text)
    count-=1
    if count!=0:
        urlmaker(pl,count)
    count
    if count==1:
        print("    "+i)
        print("\n")
    else:
        for x in pl:
            print(x)
        print("  "+ i)
    Master.append(pl)                
```
